### 4-HBA, PCA and 2,5 PDCA

```python
import pandas as pd

directory = "C:/Users/ba03/OneDrive/Desktop/biome/" #<- change file directory/folder here 
xls_file = pd.ExcelFile(directory + "xlstest.XLS")#<-specific file name to be read
xls_file = pd.read_excel(xls_file)

dfdict = {}
samplenames = []

def extractsamples1(filename, HBA_window, PCA_window, PDCA25_window, xls=xls_file):
    dfdict = {}
    samplenames = []
    for sheetname in xls.sheet_names:
        pdfile = pd.read_excel(xls, sheet_name=sheetname)
        samplename = pdfile.loc[2, 'Unnamed: 6']
        if samplename in samplenames:
            samplename = str(samplename) + 'i'
        
        samplenames.append(samplename)
        
        try:
            if 'blank' in samplename or 'water' in samplename.lower(): #sheets in file that the code will avoid in data output
                continue
    
        except TypeError:
            continue
    
        signal260 = 0
        signal280 = 0
        
        for i, row in enumerate(pdfile['Unnamed: 2']):
        
            if row == 'DAD1 C, Sig=260,4 Ref=off':
                signal260 = i
            
            elif row == 'DAD1 D, Sig=280,4 Ref=off':
                signal280 = i
            
            else:
                continue
        
        datalist = [None, None, None, None]
     
        for i, rt260 in enumerate(pdfile.loc[signal260+2:signal280-1, 'Area Percent Report']):
            if float(rt260) >= HBA_window[0] and float(rt260) <= HBA_window[1]: #4-HBA window
                datalist[0] = float(pdfile.loc[i+signal260+2, 'Unnamed: 7'])
                
            if float(rt260) >= PCA_window[0] and float(rt260) <= PCA_window[1]: #PCA window
                datalist[1] = float(pdfile.loc[i+signal260+2, 'Unnamed: 7'])
                
        for i, rt280 in enumerate(pdfile['Area Percent Report'].iloc[signal280+2:]):
            if float(rt280) >= PDCA24_window[0] and float(rt280) <= PDCA24_window[1]: #PDCA window
                datalist[2] = float(pdfile.loc[i+signal280+2, 'Unnamed: 7'])
                
            if float(rt280) >= PDCA25_window[0] and float(rt280) <= PDCA25_window[1]: #PDCA window
            datalist[3] = float(pdfile.loc[i+signal280+2, 'Unnamed: 7'])
       
        dfdict[samplename] = datalist
    
    alldata = pd.DataFrame(dfdict, index=['4-HBA', 'PCA', '2,4 PDCA', '2,5 PDCA'])
    formatedtable = alldata.transpose()
    formatedtable.to_excel(directory + filename, index=True)


extractsamples1(filename= 'jupytertest.xlsx', HBA_window=[9.2, 9.3], PCA_window=[7.45, 8.0], PDCA24_window[1.5,1.6], PDCA25_window=[1.9, 2.1])#<- enter windows here
   

```

### Glucose

```python
glu_file = pd.ExcelFile(directory + "15082024_BRX12_Glucose.XLS")#<- specific file name to be read

def extractglucose(filename, glucose_window, xls_file=glu_file):
    dfdict = {}
    samplenames = []
    for sheetname in xls_file.sheet_names:
        pdfile = pd.read_excel(xls_file, sheet_name=sheetname)
        samplename = pdfile.loc[2, 'Unnamed: 6']
        if samplename in samplenames:
            samplename = samplename + 'i'
        
        samplenames.append(samplename)
        
        try:
            if 'Blank' in samplename or 'blank' in samplename or 'water' in samplename: #sheets in file that the code will avoid in data output
                continue
    
        except TypeError:
            continue
     
        for i, rt in enumerate(pdfile.loc[21:None, 'Area Percent Report']):
            if float(rt) >= glucose_window[0] and float(rt) <= glucose_window[1]: #glucose window
                dfdict[samplename] = float(pdfile.loc[i + 21, 'Unnamed: 7'])
    
     
    
    alldata = pd.DataFrame(dfdict, index=['Glucose Area'])
    formatedtable = alldata.transpose()
    
    formatedtable.to_excel(directory + filename, index=True)

extractglucose(filename='glucosetestjupyter.xlsx', glucose_window=[14.4, 14.6])
```
